```python
def loadpredictorandtarget_data(df, predictor_variable_list, target_variable_list):
    predictor_data, target = [], []
    for row in df[predictor_variable_list].values:
        predictor_data.append(row)
    for row in df[target_variable_list].values:
        target.append(row[0])
    return predictor_data, target
```

```python
#How did you select variables to be included in the model?
#From the data we can observe that all  the fraud transactions are done for the 
#payment type 'TRANSFER','CASH OUT' only.
```

```python
#so we can observe fraud is only commited for transaction types 'Transfer','Cash Out,
allowabletransactiontypes=['TRANSFER','CASH_OUT']
```

```python
#as i only want rows with these types 
df = df.loc[df['type'].isin(allowabletransactiontypes)]
```

```python
#What are the key factors that predict fraudulent customer
#'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest'
```

```python
targetvar = ['isFraud']
predictorvar=['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
usepredictordata, target= loadpredictorandtarget_data(df, predictorvar, targetvar)
```

```python
x_train, x_test, y_train, y_test = ms.train_test_split(usepredictordata,target, test_size=0.25, random_state=None)
```

```python
# so we use logistic regression and confusion matrix to train, test and represent our data.
```

```python
logisticRegr = LogisticRegression()  
logisticRegr.fit(x_train, y_train)
```

```python
#making predictions on the remaining data
predictions = logisticRegr.predict(x_test)
```

```python
confusion_matrix_log = confusion_matrix(y_test, predictions)
print('confusion matrix\n', confusion_matrix_log)
```

```python
#lets check accuracy
score = logisticRegr.score(x_test, y_test)
print('Score: ', score)
```
